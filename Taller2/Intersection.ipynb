### Ejemplo:

Para encontrar la solución a la ecuación trascendental:

$${\color{black}{\sqrt[x]{x}=20\ \text{sinc}(x)}}\tag{3}$$

donde $\text{sinc}(x) \equiv \frac{\sin(x)}{x}$, se calcula los ceros de la resta:

$$\sqrt[x]{x}-20\ \text{sinc}(x)=0\tag{4}$$

es decir, si llamamos:

$${\color{blue}{f(x)=\sqrt[x]{x}}}, \qquad {\color{red}{g(x)= \text{sinc}(x)}},$$

queremos encontrar 
$${\color{blue}{f(x)}}-{\color{red}{g(x)}}=0\tag{5}$$

Así el problema se reduce a hallar los ceros de la ecuación ([4](#mjx-eqn-4)) usando la forma ([5](#mjx-eqn-5)).

````python
import numpy as np
import matplotlib.pyplot as plt

# To improve resolution of image in jupyter notebook:
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
````

````python
x = np.linspace(0.001,7,1000)

f = x**(1/x)
g = 20*np.sinc(x)
````

#### Hallamos los ceros de la función

````python
# Para encontrar los ceros de la función
index = np.argwhere(np.diff(np.sign(f - g))).flatten()
````

- ```f - g```: Resta entre las dos funciones.

- ```np.sign(f - g)```: Halla el signo de  ```f - g```  (```+1``` si es positivo, ```-1``` si es negativo).

- ```np.diff(np.sign(f - g))```: Realiza la diferencia entre una partición y la siguiente de la función signo. Si llegase a encontrar un cambio de signo, lo asigna en este caso como ```2``` ó ```-2```.

- ```np.argwhere(np.diff(np.sign(f - g)))```: Localiza los elementos (particiones) que no sean cero.

- ```np.argwhere(np.diff(np.sign(f - g))).flatten()```: Convierte los elementos en un arreglo (array) de datos.

````python
print("Los índices son:",index)
print("Los valores x:  ",x[index])
````

````python
plt.plot(x,g,label=r'$g(x)=20\ \sinc(x)$')
plt.plot(x,f,label=r'$f(x)=x^{1/x}$')

plt.plot(x[index], g[index], 'ro')
plt.grid()
plt.legend(fontsize=11)
````

### Ejemplo 2:

Vamos a hallar los valores de $x$ que safisfagan: 

\begin{align}
m&=\dfrac{x}{x^3}\\
m~x^3&=x\\
\color{blue}{m~x^3}-\color{red}{x}&=0\\
(m~x^2-1)x&=0
\end{align}
Cuyas soluciones son:
$$x=\pm\sqrt{\frac{1}{m}},\qquad x=0$$

Hacemos:

$${\color{blue}{f(x)=m~x^3}}, \qquad {\color{red}{g(x)= x}},$$

````python
x = np.linspace(0.01,1.5,1000)
m = np.arange(1,10,0.1)

arreglo = [] # lista para agregar valores para cada "m"
for MASAS in m:
    f = MASAS*x**3
    g = x
    index = np.argwhere( np.diff( np.sign(f - g) ) ).flatten()
    # Agregamos elementos al final de la lista "arreglo":
    arreglo.append(x[index][0])  
    #print("%.2f"%x[index][0],end=" ")
````

````python
plt.plot(m,arreglo,c='b')
plt.xlabel(r"$m$",fontsize=13)
plt.ylabel(r"$x$",fontsize=13)
plt.grid()
````
